```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/employee_data.csv")
```

```python
print("Employee Salary Dataset:")
print(df.head())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Summary Statistics
print("\nSummary Statistics:")
print(df.describe())

# Salary Distribution Plot
plt.figure(figsize=(8, 5))
sns.histplot(df['Salary'], bins=10, kde=True, color='skyblue')
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()

# Salary vs. Experience
plt.figure(figsize=(8, 5))
sns.barplot(x=df['Department'], y=df['Salary'], estimator=sum, ci=None, palette='pastel')
plt.title("Salary vs. Experience")
plt.xlabel("Experience (Years)")
plt.ylabel("Salary")
plt.show()

# Average Salary by Department
plt.figure(figsize=(8, 5))
sns.barplot(x=df['Department'], y=df['Salary'], estimator=sum, ci=None, palette='pastel')
plt.title("Total Salary by Department")
plt.xlabel("Department")
plt.ylabel("Total Salary")
plt.show()

# Insights
print("\nInsights:")
print("- The salary distribution is approximately normal with a slight right skew.")
print("- Employees with more experience tend to have higher salaries.")
print("- The Finance department has the highest total salary allocation.")
print("- Checking for missing values and overall summary helps in better understanding of the dataset.")
```
